### Accepted name references

(These are taxon concepts)

```python
# References
data = []
for index, row in df_accepted.iterrows():
    print(row['ScientificName'])
    name_id = str(int(row['NameId']))
    url = 'http://services.tropicos.org/Name/' + name_id + '/References'
    query = {'apikey': apikey.apikey, 'format': 'json'}

    r = requests.get(url, query)
    response = r.json()
    for item in response:
        item['AcceptedName'] = {
            'NameId': row['NameId'],
            'ScientificName': row['ScientificName'],
            'ScientificNameWithAuthors': row['ScientificNameWithAuthors']
        }
        data.append(item)

df_nref = pd.json_normalize(data)

# Use only the references in the list
df_name_references = df_nref[df_nref['Reference.ReferenceId'].isin(references)]
df_name_references
```

### Synonyms

```python
#synonyms
data = []
for index, row in df_accepted.iterrows():
    print(row['ScientificName'])
    name_id = str(int(row['NameId']))
    url = 'http://services.tropicos.org/Name/' + name_id + '/Synonyms'
    query = {'apikey': apikey.apikey, 'format': 'json'}

    r = requests.get(url, query)
    data = data + r.json()

df_syn = pd.json_normalize(data)
df_synonym = df_syn[df_syn['Reference.ReferenceId'].isin(references)]
df_synonym

```

```python
accepted_names = df_accepted['NameId'].astype(int).to_list()
synonyms = df_synonym.drop_duplicates(subset=['SynonymName.NameId'])['SynonymName.NameId'].astype(int).to_list()
all_names = list(set(accepted_names + synonyms))

data = []
for id in all_names:
    url = 'http://services.tropicos.org/Name/' + str(id)
    query = {'apikey': apikey.apikey, 'format': 'json'}

    r = requests.get(url, query)
    body = r.json()
    print(body['ScientificNameWithAuthors'])
    data.append(body)

df_names = pd.DataFrame(data)
```

### List taxon concepts

```python
df_synonym.drop_duplicates(subset=['AcceptedName.NameId', 'Reference.ReferenceId', 'SynonymName.NameId'], inplace=True)
df_synonym.sort_values(by=['AcceptedName.ScientificName', 'Reference.TitlePageYear', 'SynonymName.ScientificName'], inplace=True)

df_name_references.drop_duplicates(subset=['AcceptedName.NameId', 'Reference.ReferenceId'], inplace=True)
df_name_references.sort_values(by=['AcceptedName.ScientificName', 'Reference.TitlePageYear'], inplace=True)
```

```python
def get_name(id):
    dict = df_names[df_names['NameId'] == id].to_dict(orient='records')[0]
    name = {
        'nameString': None,
        'authorship': None,
        'proto': None,
        'status': None,
        'basionym': None
    }
    name['nameString'] = dict['ScientificName'].replace('fo.', 'f.')
    if isinstance(dict['Author'], str):
        if isinstance(dict['BasionymAuthor'], str):
            authorship = '(' + dict['BasionymAuthor'] + ') ' + dict['Author']
        else:
            authorship = dict['Author']
        name['authorship'] = authorship.replace('. ', '.').replace('.&', '. &').replace('.ex', '. ex')
        name['proto'] = dict['NamePublishedCitation'].strip()
        if dict['NomenclatureStatusName'] in ['Illegitimate', 'Invalid']:
            name['status'] = dict['NomenclatureStatusName'].lower()
        name['basionym'] = int(dict['BasionymId']) if isinstance(dict['Basionym'], str) else None
    return name

get_name(35158613)
```

```python
def get_synonyms(name_id, ref_id):
    return df_synonym[(df_synonym['AcceptedName.NameId'] == name_id) 
            & (df_synonym['Reference.ReferenceId'] == ref_id)]['SynonymName.NameId'].astype(int).to_list()

get_synonyms(35158613, 9020903)
```

```python
with open('tropicos-malesian-dicranoloma.md', 'w') as f:
    for index, row in df_name_references.iterrows():
        an = get_name(row['AcceptedName.NameId'])
        
        f.write('**' + an['nameString'] + '** ' + (an['authorship'] if an['authorship'] else '') + ' [sec. ' + references[int(row['Reference.ReferenceId'])] + ']\n\n')
        if an['basionym']:
            b = get_name(an['basionym'])
            f.write('  ≡ *' + b['nameString'] + '* ' + b['authorship'] + '\n\n')
        syns = get_synonyms(int(row['AcceptedName.NameId']), int(row['Reference.ReferenceId']))
        if len(syns):
            for syn in syns:
                s = get_name(syn)
                if s['status'] == 'invalid':
                    stat = ', *nom. inval.*'
                elif s['status'] == 'illegitimate':
                    stat = ', *nom. illeg.*'
                else:
                    stat = ''
                f.write('  = *' + s['nameString'] + '* ' + s['authorship'] + stat + '\n\n')
        f.write('<br/>\n\n')
```

### TurTLe

```python
df_names.sort_values(by='ScientificName', inplace=True)

prefix = """@prefix : <http://rs.tdwg.org/tcs/terms/> .
@prefix dcterms: <http://purl.org/dc/terms/> .
@prefix dwc: <http://rs.tdwg.org/dwc/terms/> .


"""
with open('tropicos-malesian-dicranoloma.ttl', 'w') as f:
    f.write(prefix)

    for index, row in df_name_references.iterrows():
        f.write('<https://www.tropicos.org/reference/' + str(int(row['Reference.ReferenceId'])) + '#name-' + str(int(row['AcceptedName.NameId'])) + '> a :TaxonConcept ;\n')
        f.write('    dcterms:title "' + row['AcceptedName.ScientificName'].replace('fo.', 'f.') + ' sec. ' + references[int(row['Reference.ReferenceId'])] + '" ;\n')
        f.write('    :accordingTo <https://www.tropicos.org/reference/' + str(int(row['Reference.ReferenceId'])) + '> ;\n')

        
        synonyms = df_synonym[(df_synonym['AcceptedName.NameId'] == row['AcceptedName.NameId']) & (df_synonym['Reference.ReferenceId'] == row['Reference.ReferenceId'])]
        if len(synonyms):
            f.write('    :acceptedName <https://www.tropicos.org/name/' + str(int(row['AcceptedName.NameId'])) + '> ;\n')
            syns = []
            for i, syn in synonyms.iterrows():
                syns.append('<https://www.tropicos.org/name/' + str(int(syn['SynonymName.NameId'])) + '>')
            f.write('    :synonym ' + ' ,\n        '.join(syns) + ' .\n\n')

        else:
            f.write('    :acceptedName <https://www.tropicos.org/name/' + str(int(row['AcceptedName.NameId'])) + '> .\n\n')


    for index, row in df_names.iterrows():
        f.write('<https://www.tropicos.org/name/' + str(int(row['NameId'])) + '> a :TaxonName ;\n')
        f.write('    dwc:scientificName "' + row['ScientificName'].replace('fo.', 'f.') 
                + row['ScientificNameWithAuthors'][len(row['ScientificName']):]
                .replace('. ', '.').replace('.&', '. &').replace('.ex', '. ex') + '" ;\n')

        if isinstance(row['Author'], str):
            f.write('    :taxonNameString "' + row['ScientificName'] + '" ;\n')
            if isinstance(row['Basionym'], str):
                f.write('    dwc:scientificNameAuthorship "(' + row['BasionymAuthor']
                        .replace('. ', '.').replace('.&', '. &') + ')' 
                        + row['Author'].replace('. ', '.').replace('.&', '. &').replace('.ex', '. ex') + '" ;\n')

                bas = df_names[df_names['NameId'] == row['BasionymId']].to_dict(orient='records')[0]
                f.write('    :basionym <https://www.tropicos.org/name/' + str(int(bas['NameId'])) + '> ;\n')

            else:
                f.write('    dwc:scientificNameAuthorship "' + row['Author'].replace('. ', '.').replace('.&', '. &').replace('.ex', '. ex') + '" ;\n')
                f.write('    dwc:namePublihedIn "' + row['NamePublishedCitation'].strip() + '" ;\n')

            if row['NomenclatureStatusName'] in ['Legitimate', 'No opinion']:
                status = '<http://rs.gbif.org/vocabulary/gbif/nomenclatural_status/legitimate>'
            elif row['NomenclatureStatusName'] == 'Illegintimate':
                status = '<http://rs.gbif.org/vocabulary/gbif/nomenclatural_status/illegitimum>'
            elif row['NomenclatureStatusName'] == 'Invalid':
                status = '<http://rs.gbif.org/vocabulary/gbif/nomenclatural_status/invalidum>'

            f.write('    :nomenclaturalStatus ' + status + ' .\n\n')

        else:
            f.write('    :taxonNameString "' + row['ScientificName'] + '" .\n\n')
```
